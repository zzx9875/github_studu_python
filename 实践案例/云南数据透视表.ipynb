```python
import  pandas as pd
df = pd.read_excel('./Desktop/2019_云南.xlsx',sheet_name='2019',header=4)
df
```

```python
df[['省份','城市']] = df['云南城市'].str.split('-',expand=True)
df
```

```python
a = df['销售金额'].groupby(df['城市']).sum()
a
```

```python
df_sum = df.groupby('城市').agg({'销售(台)':sum,
                           '退回(台)':sum,
                           '退款金额':sum,
                           '销售金额':sum,
                           '业绩任务数':sum,
                           '提成金额':sum})
df_sum
```

```python
df_sum.loc['总计'] = df_sum.apply(lambda x: x.sum())
df_sum
```
